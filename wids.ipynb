```python
all_data2 = all_data.drop(columns=['direction_max_wind_speed','direction_peak_wind_speed','max_wind_speed','days_with_fog','id','site_eui'])
all_data2['energy_star_rating'] = all_data2['energy_star_rating'].fillna(all_data2['energy_star_rating'].mean())
all_data2['year_built'] = all_data2['year_built'].fillna(all_data2['year_built'].mean())
all_data2
```

```python
final_columns = ['energy_star_rating','facility_type','floor_area','year_built','State_Factor','ELEVATION']

all_data3 = all_data2[final_columns]
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

all_data3['State_Factor'] = le.fit_transform(list(all_data3['State_Factor']))
all_data3 = pd.get_dummies(all_data3)
all_data3
```

```python
train2 = all_data3[:len(train)]
test2 = all_data3[len(train):]
```

```python
from sklearn.model_selection import cross_validate
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_jobs = -1)

dt_cv_scores = pd.DataFrame(cross_validate(rf, train2, train['site_eui'], cv=5, return_train_score=True, scoring = 'neg_mean_squared_error'))

dt_cv_scores
```

```python
rf.fit(train2, train['site_eui'])
```

```python
from sklearn.metrics import mean_squared_error

result_r = rf.predict(test2)

mean_squared_error(train['site_eui'].tolist(), rf.predict(train2))**(1/2)
```

```python
feature_importance = pd.DataFrame({'feature': train2.columns, 'importance': rf.feature_importances_}).sort_values(by = 'importance', ascending = False)
feature_importance
```

```python
sub['site_eui'] = result_r
sub.to_csv('result.csv', index = 0)
```
